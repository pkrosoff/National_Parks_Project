```python
import pandas as pd

# read in data
park_info = pd.read_csv('../data/np_data.csv')
park_atten_pre79 = pd.read_csv('../data/npvisitation_1904_1978.csv')
park_atten_post79 = pd.read_csv('../data/npvisitation_1979_2019.csv')

# splitting, reformatting area types
park_info[['area_acres','area_km2']] = park_info.area_km2.str.split("(",expand=True)
area_acres = park_info['area_acres'].replace({' acres':'',',':''}, regex=True)
area_acres = area_acres.str.strip().to_frame().astype('float')
park_info['area_Acres'] = area_acres

# finding km2 and mi2 with math
park_info['area_kM2'] = round(park_info['area_Acres']*0.00404686,2)
park_info['area_miles2'] = round(park_info['area_Acres']*0.0015625,2)
park_info = park_info.drop(columns=['area_km2', 'area_acres']).rename(columns={'area_Acres':'area_acres', 'area_kM2':'area_km2'})
park_info
```
